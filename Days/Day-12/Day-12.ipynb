```python
df = spark.table("workspace.ecommerce.events_delta")
```

```python
#Analyze job Runtime
import time

start = time.time()
df.groupBy("event_type").count().show()
end = time.time()

print("Execution Time:", end - start, "seconds")
```

Noted Observations:
Repeated execution showed reduced runtime due to Spark executor warm-up and Delta metadata caching.
first run:0.988
second run:0.750
That’s measured optimization.

```python
#Reduce Unnecessary Actions
#Bad pattern:
df.count()
df.groupBy("event_type").count().show()
df.count()    # each count() triggers a full scan
```

```python
#Better pattern:
grouped = df.groupBy("event_type").count()
grouped.show()  #only one action executed
```

```python
df.groupBy("event_type").count().explain()
```
