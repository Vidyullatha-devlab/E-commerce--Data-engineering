```python
events_df = spark.table("workspace.ecommerce.events_delta")
```

```python
new_data = events_df.limit(5)

new_data.write.format("delta") \
    .mode("append") \
    .saveAsTable("workspace.ecommerce.events_delta")
```

```python
current_df = spark.table("workspace.ecommerce.events_delta")

old_df = spark.read.format("delta") \
    .option("versionAsOf", 1) \
    .table("workspace.ecommerce.events_delta")

print("Current count:", current_df.count())
print("Old count:", old_df.count())
```
